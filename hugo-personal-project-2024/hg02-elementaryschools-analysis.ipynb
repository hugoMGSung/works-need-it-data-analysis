## 부산 초등학교 현황

### 분석데이터

학사/학생 > 학년별.학급별 학생수 API로 진행

#### 메타정보
- 분류체계 : 학사/학생 > 학년별·학급별 학생수
- 요청 URL : http://www.schoolinfo.go.kr/openApi.do
- 제공방식 : REST
- 데이터포맷 : JSON

- 요청인자
	- apiKey STRING(필수) 인증키
	- apiType STRING(필수) Api종류 09
	- pbanYr NUMBER(필수) 공시연도 yyyy(2020) 최근 3년간 정보만(2024년 안됨)
	- schulKndCode STRING(필수) 학교급구분 02:초등 03:중등 04:고등 05:특수 06:그외 07:각종

- 출력인자
	- 181개 생략

- 아래와 같은 방식으로 URL을 요청해야
	- https://www.schoolinfo.go.kr/openApi.do?apiKey=c4397fe308a241feb79cc7f7c58c0282&apiType=09&schulKndCode=02&pbanYr=2023


### 라이브러리 Import 및 URL

```python
import requests
import json
import pprint
import certifi
import pandas as pd
```

```python
url = 'https://www.schoolinfo.go.kr/openApi.do?apiKey=c4397fe308a241feb79cc7f7c58c0282&apiType=09&schulKndCode=02&pbanYr=2023'
```

```python
response = requests.get(url)
contents = response.text
```

```python
## 위의 문제 port=443): Max retries exceeded with url 해결. 
response = requests.get(url, verify=False)
contents = response.text
```

```python
json_object = json.loads(contents)
```

```python
json_object
```

```python
with open('result.json', 'w', encoding='utf-8') as f:
    json.dump(json_object, f, ensure_ascii=False, indent=4)
```

```python
json_object['list']
```

```python
df = pd.DataFrame(json_object['list'])
```

```python
df
```

```python
df.to_csv('korean_elementry_status_2023.csv')
```
