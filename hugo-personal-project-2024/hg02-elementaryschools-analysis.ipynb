학사/학생 > 학년별.학급별 학생수 API로 진행

#### 메타정보
- 분류체계 : 학사/학생 > 학년별·학급별 학생수
- 요청 URL : http://www.schoolinfo.go.kr/openApi.do
- 제공방식 : REST
- 데이터포맷 : JSON

- 요청인자
	- apiKey STRING(필수) 인증키
	- apiType STRING(필수) Api종류 09
	- pbanYr NUMBER(필수) 공시연도 yyyy(2020) 최근 3년간 정보만(2024년 안됨)
	- schulKndCode STRING(필수) 학교급구분 02:초등 03:중등 04:고등 05:특수 06:그외 07:각종

- 출력인자
	- 181개 생략

- 아래와 같은 방식으로 URL을 요청해야
	- https://www.schoolinfo.go.kr/openApi.do?apiKey=c4397fe308a241feb79cc7f7c58c0282&apiType=09&schulKndCode=02&pbanYr=2023


### 라이브러리 Import 및 URL

- requests(Python HTTP for Humans) : !pip install requests

```python
import requests
import json
import pprint
import certifi
import pandas as pd
```

```python
url = 'https://www.schoolinfo.go.kr/openApi.do?apiKey=c4397fe308a241feb79cc7f7c58c0282&apiType=09&schulKndCode=02&pbanYr=2023'
```

```python
## 오류발생
## SSLError: HTTPSConnectionPool(host='www.schoolinfo.go.kr', port=443): Max retries exceeded with url: /openApi.do?
response = requests.get(url)
contents = response.text
```

```python
## 위의 문제 port=443): Max retries exceeded with url 해결. 
response = requests.get(url, verify=False)
contents = response.text
```

```python
json_object = json.loads(contents)
```

```python
## 되도록이면 확인하지 말것/ 아래에서 저장후 확인 요
# json_object
```

```python
with open('result.json', 'w', encoding='utf-8') as f:
    json.dump(json_object, f, ensure_ascii=False, indent=4)
```

```python
## json을 pandas로 변환
df = pd.DataFrame(json_object['list'])
```

```python
df.head()
```

```python
# 초등학교 정보 csv로 저장
df.to_csv('korean_elementry_status_2023.csv')
```

```python
# 엑셀보다 csv가 편하다. 저장시간 및 파일 사이즈 등...
df.to_excel('korean_elementry_status_2023.xlsx')
```

#### 두번째 방식

	- 다운로드로 2024_busan_elementry_stat.csv 다운로드 
	

```python
df_busan = pd.read_csv('./2024_busan_elementry_stat.csv')
df_busan.head()
```

```python
# 필요없는 컬럼 삭제
## 위에 나오는 컬럼명을 볼 수 있는 정도까지만 삭제, 이를 반복
df_busan = df_busan.drop(['시도교육청', '교육지원청',  '학교급코드', '제외사유', '수업교원 1인당 학생수'], axis=1)
df_busan.head()
```

```python
df_busan = df_busan.drop(['1학년 학급당 학생수', '2학년 학급당 학생수', '3학년 학급당 학생수', '4학년 학급당 학생수'], axis=1)
df_busan.head()
```

```python
df_busan = df_busan.drop(['5학년 학급당 학생수', '6학년 학급당 학생수'], axis=1)
df_busan.head()
```

```python
df_busan = df_busan.drop(['특수학급 학급수', '특수학급 학생수', '특수학급 학급당 학생수', '순회학급 학급수', '순회학급 학생수'], axis=1)
df_busan.head()
```

```python
df_busan = df_busan.drop(['순회학급 학급당 학생수', '학급당 학생수(계)'], axis=1)
df_busan.head()
```

```python
df_busan = df_busan.drop(['1학년 학급수', '2학년 학급수', '3학년 학급수', '4학년 학급수', '5학년 학급수', '6학년 학급수'], axis=1)
df_busan.head()
```

```python
df_busan = df_busan.drop(['학급수(계)'], axis=1)
df_busan.head()
```

```python
# 모든 컬럼을 삭제하고 다시 저장
df_busan.to_csv('./2024_busan_elementry_final.csv')
```

```python
df_busan.tail()
```

```python
## 동일한 문자 제거
df_busan['지역'] = df_busan['지역'].str.replace('부산광역시', '')
df_busan.head()
```

```python
## 학생수(계) 컬럼명을 총 학생수로 변경
df_busan.rename(columns={'학생수(계)':'총학생수'}, inplace=True)
df_busan.head()
```
