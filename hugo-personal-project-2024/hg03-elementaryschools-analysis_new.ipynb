### 부산 초등학교 정보 로드

```python
import pandas as pd
import numpy as np

import folium
import json

from folium import Marker
from folium.plugins import MarkerCluster
```

```python
df_busan_elementry = pd.read_excel('./2024년_부산_초등학교_학생수_최종정리.xlsx')
```

```python
df_busan_elementry.head()
```

### 각 학교별로 위치값 찾아서 매핑하기
- 아래 URL에 존재
    - https://www.data.go.kr/data/15021148/standard.do 

```python
# CSV파일 로드 하기
df_school_info = pd.read_csv('./전국초중등학교위치표준데이터.csv', encoding='EUC-KR')
```

```python
df_school_info.tail(2)
```

#### 시도교육청명 이 '부산'으로 시작하는 정보만 발췌

```python
df_school_info['시도교육청명'].str.contains('부산')
```

```python
df_busan_school_info = df_school_info[df_school_info['시도교육청명'].str.contains('부산')]
```

```python
df_busan_school_info.tail(2)
```

```python
df_busan_school_info.head(2)
```

#### 학교급구분 이 '초등학교' 인 곳 필터링

```python
df_busan_school_info['학교급구분'].str.contains('초등학교')
```

```python
df_busan_elementry_info = df_busan_school_info[df_busan_school_info['학교급구분'].str.contains('초등학교')]
```

```python
df_busan_elementry_info.tail()
```

#### 학교 ID, 학교명, 설립일자, 소재지도로명주소, 위도, 경도 만 두고 나머지 삭제

```python
df_busan_elem_last = df_busan_elementry_info[['학교ID', '학교명', '설립일자', '소재지도로명주소', '위도', '경도']]
```

```python
df_busan_elem_last.tail()
```

#### 두개 데이터프레임 합치기

```python
df_last_of_busan = pd.merge(df_busan_elementry, df_busan_elem_last, how='inner', left_on='학교명', right_on='학교명')
```

```python
df_last_of_busan.to_excel('./2024년_부산_초교_정보최종.xlsx')
```

```python
df_last_of_busan.tail()
```

#### 지도 표시

```python
busan_center = [35.1379222, 129.05562775]
map = folium.Map(busan_center, tiles="OpenStreetMap", zoom_start=11)
map
```

```python
geo_str = json.load(open('./busan_gu.json', encoding='utf-8'))

folium.GeoJson(geo_str, name='json_data').add_to(map)
map
```

#### 각 학교 위경도로 지도에 표시

```python
for _, i in df_last_of_busan.iterrows():
    # print(i['위도'], i['경도']) ## 총학생수, 학교명, 설립구분, 소재지도로명주소
    Marker(location= [i['위도'], i['경도']],
           popup=f"{i['학교명']} / {i['총학생수']}명",
           icon=folium.Icon(color='blue')).add_to(map)
```
