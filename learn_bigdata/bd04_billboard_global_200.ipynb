```python
# 크롬드라이버 실행
driver = webdriver.Chrome() 

#크롬 드라이버에 url 주소 넣고 실행
driver.get(url)

# 페이지가 완전히 로딩되도록 3초동안 기다림
time.sleep(3)
```

```python
search_box = driver.find_elements(By.CLASS_NAME, 'o-chart-results-list-row-container')
```
