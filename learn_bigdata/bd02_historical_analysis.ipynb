```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({"name":"John Doe","value":np.random.randn(30)*1000,"date":pd.period_range("2019/10/1","2019/10/30")})
```

```python
df.head()
```

```python
ls_name = ["GilDong Hong","Done Lee","Robert Downey Jr.","Leonardo Dicaprio","Joseph Gordon-Levitt"]
for name in ls_name:
    df_ = pd.DataFrame({"name":name,"value":np.random.randn(30)*1000,"date":pd.period_range("2019/10/1","2019/10/30")})
    df = pd.concat([df,df_])
df
```

```python
df.to_excel(r"./data/demo.xlsx",index=False)
```

```python
rng = np.random.RandomState(1)
df = pd.DataFrame(rng.randint(10,100,size=(20,10))).add_prefix("y")
# 20개의 행, 10개의 데이터 열 생성 및 레이블 추가
print(df)
df_top = df.cumsum(axis=0) # 열별 합계
print(df_top)
df_right = df.cumsum(axis=1)
print(df_right)
```

```python
df = pd.DataFrame({"name":"John Doe","value":np.random.randn(30)*1000,"date":pd.period_range("2019/10/1","2019/10/30")})
```

```python
df["value"]= df["value"].cumsum(axis=0)
df
```

```python
ls_name = ["GilDong Hong","Done Lee","Robert Downey Jr.","Leonardo Dicaprio","Joseph Gordon-Levitt"]
for name in ls_name:
    df_ = pd.DataFrame({"name":name,"value":np.random.randn(30)*1000,"date":pd.period_range("2019/10/1","2019/10/30")})
    df_["value"]= df_["value"].cumsum(axis=0)
    df = pd.concat([df,df_])
df
```

```python
df.to_excel(r"./data/demo_acc.xlsx",index=False)
```
