### 판다스 데이터 전처리 학습

- 데이터 분석 전에 가장 중요한 작업 - 데이터 전처리
    - 잘못된 데이터를 없애는 작업
    - 전체 프로젝트에 50% 정도 기간이 소요

```python
import pandas as pd
```

```python
data = {
    '이름': ['홍길동', '홍길순', '성유고', '박애슐'],
    '국어': [80, 100, 10, 90],
    '영어': [50, 100, 99, 100],
    '수학': [90, 100, 10, 40],
    '미술': [100, 100, 99, None]  # Python NULL
}

df = pd.DataFrame(data=data)
```

```python
# NaN : Not a Number == None
df
```

```python
df.info()
```

```python
df.describe()
```

#### 결측치 확인, 결측치 제거
- NaN(None) 값을 찾아서 제거하는 것!

```python
# 결측치 갯수 합산
df.isnull().sum()
```

```python
df.isna().sum()
```

```python
df2 = df.copy()
```

#### 결측치 제거 1 - dropna
- 결측치가 존재하는 레코드 삭제

```python
## 결측치가 있는 데이터 제거, inplace=True 하면 완전제거
### 실무에선 사용하면 안되는 기능!
df2.dropna(inplace=True)
```

```python
df2
```

```python
df3 = df.copy()
```

#### 결측치를 채우기
- 결측치를 일괄적인 값으로 변경

```python
# df3['미술'] = df3['미술'].fillna(70)
```

```python
df3['미술'] = df3['미술'].fillna(df3['미술'].mean())
```

```python
df3
```

```python
df['미술'].min()
```

##### 데이터프레임에서 값을 변경할 때 중요점
1. 바뀐 값을 다시 원본에 할당!
2. 함수 파라미터 중 inplace=True 기능을 사용하면 변경이 반영

```python
df_titanic['pclass'] = df_titanic['pclass'].astype('int8')
```

```python
df_titanic.info()
```

```python
## str(문자열로 변환) -> object 타입
df_titanic['pclass'] = df_titanic['pclass'].astype('str')
```

```python
df_titanic.info()
```
