### Seaborn 데이터 관련

#### 시본에 들어있는 샘플데이터

```python
import seaborn as sns
```

```python
## 시본 설정
sns.set_theme(style='darkgrid', color_codes=True)
```

```python
## 붓꽃 데이터
iris_data = sns.load_dataset(name='iris')
iris_data.tail()
```

```python
## 타이타닉 데이터
titanic_data = sns.load_dataset('titanic')
titanic_data.tail()
```

```python
## 팁 데이터
tip_data = sns.load_dataset('tips')
tip_data.tail()
```

```python
## 택시요금 데이터
taxifare_data = sns.load_dataset('taxis')
taxifare_data.tail()
```
