##### 구글 플러그인 SelectorsHub 사용 권장

- XPath 확인시 가장 좋음

```python
table_data = driver.find_elements(By.XPATH, "//table[@id='jsGrid_MDCSTAT021_0']//td")
len(table_data)
```

```python
for data in table_data:
    print(data.text)
```

```python
5 % 6
```

```python
## 데이터를 담을 수 있는 방법으로 담으면 됨
## 총 컬럼이 6개 이므로
lst_samsung_stocks = []
cur_stock = {}

for i, data in enumerate(table_data):
    #print(i, data.text)
    if i % 6 == 0: cur_stock['Date'] = data.text
    if i % 6 == 1: cur_stock['LastPrice'] = data.text
    if i % 6 == 2: cur_stock['Closing'] = data.text
    if i % 6 == 3: cur_stock['UpDown'] = data.text
    if i % 6 == 4: cur_stock['Volumn'] = data.text
    if i % 6 == 5: 
        cur_stock['TradeCost'] = data.text
        lst_samsung_stocks.append(cur_stock)
        cur_stock = {}       
     
```

```python
lst_samsung_stocks
```

```python
## 데이터프레임으로 바꾸고자 하면
stocks = {}

dates = []
lastprices = []
closings = []
updowns = []
volumns = []
tradecosts = []

for i, data in enumerate(table_data):
    if i % 6 == 0: dates.append(data.text)
    if i % 6 == 1: lastprices.append(int(data.text.replace(',', '')))
    if i % 6 == 2: closings.append(int(data.text.replace(',', '')))
    if i % 6 == 3: updowns.append(data.text)
    if i % 6 == 4: volumns.append(int(data.text.replace(',', '')))
    if i % 6 == 5: tradecosts.append(int(data.text.replace(',', '')))
```

```python
stocks['일자'] = dates
stocks['종가'] = lastprices
stocks['대비'] = closings
stocks['등락률'] = updowns
stocks['거래량'] = volumns
stocks['거래대금'] = tradecosts
```

```python
## 데이터 프레임으로 변환
df_samsung_stocks = pd.DataFrame.from_dict(stocks)
```

```python
df_samsung_stocks
```

```python
df_samsung_stocks.to_csv('./results/삼성전자_종가.csv', encoding='utf-8', index=False)
```
